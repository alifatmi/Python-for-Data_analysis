```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

titan = sns.load_dataset("titanic")
```

```python
x=titan.alone
x
```

```python
# catplot
sns.catplot(x="sex",y="survived",kind="bar", data=titan)
plt.show
```

```python
# scatterplot
g =sns.FacetGrid(titan,row="sex",hue="alone")
g =(g.map(plt.scatter, "age","fare").add_legend())
```

```python
# countplot
p=sns.countplot(x='sex',data=titan,hue='alone')
```
